```
def get_date_from_log(cid_dbfs_path):
    print(f"Checking cluster id log path: {cid_dbfs_path}")
    ls = dbutils.fs.ls(cid_dbfs_path + 'driver')
    ls_driver = list(map(lambda x: x.path, ls))
    has_current_logs = False
    for logfile in ls_driver:
        if '-2021-' in logfile:
            has_current_logs = True
            return has_current_logs
    return has_current_logs
```

```
total_cluster_logs = 0
deleted_cluster_logs = 0 
for cid_info in dbutils.fs.ls('dbfs:/cluster-logs/'):
    cid_path = cid_info.path
    total_cluster_logs += 1
    if not get_date_from_log(cid_path):
        deleted_cluster_logs += 1
        print(f"Deleting log path {cid_path}\n")
        dbutils.fs.rm(cid_path, True)
```

```
print(f"Total Cluster Log Paths: {total_cluster_logs}")
print(f"Deleted Cluster Log Paths: {deleted_cluster_logs}")
```
